```python
def get_playlist_tracks(id):
    '''
    Get the tracks for a spotify playlist.
    Input: playlist ID
    Output: list of song IDs
    '''
    return [i['track']['id'] for i in sp.playlist(id,market='IN')['tracks']['items']]
```

```python
def get_the_playlists():
    '''
    Get playlist ids for a particular day
    Inputs: date in ISO-8601 format
            offfset - page number
    Output: paired playlist ids with number of tracks
    '''
    return ([i['id'],i['name'],i['tracks']['total']] for i in sp.featured_playlists(
        locale='hin_IN',
        country='IN',
        limit=50
    )['playlists']['items'])
```

```python
def get_the_playlists_by_category(category,offset):
    '''
    Get playlist ids for a particular day
    Inputs: date in ISO-8601 format
            offfset - page number
    Output: paired playlist ids with number of tracks
    '''
    return ([i['id'],i['name'],i['tracks']['total']] for i in sp.category_playlists(
        category_id=category,
        country='IN',
        limit=50,
        offset=offset
    )['playlists']['items'])
```
